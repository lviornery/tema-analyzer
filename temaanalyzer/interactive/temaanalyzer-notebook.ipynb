```python
from temaanalyzer import temafunctions as tf
import pandas as pd
```

## Raw Data Import

```python
badDf = pd.read_csv('TEMAExampleFile.txt',sep='\t',encoding_errors='ignore')
badDf.head()
```

## Combined Data Cleaning

```python
df = tf.cleanImportTemaData('TEMATotalHeader.txt')
df.head()
```

## Raw Data Import

```python
dfAlt = tf.importTemaData('TEMATotalHeader.txt')
dfAlt.head()
```

## Unit Standardization

```python
dfAlt = tf.standardizeUnits(dfAlt)
dfAlt.head()
```

## Column Format Standardization

```python
dfAlt = tf.standardizeColFormat(dfAlt)
dfAlt.head()
```

## Column Order Standardization

```python
dfAlt = tf.standardizeColOrder(dfAlt,renumber=True)
dfAlt.head()
```

## Velocity Calculation

```python
dfDiff = tf.calculateVelocity(df,'yPosition1[m]')
dfDiff
```

## Column Order Standardization with New Velocity

```python
dfDiff = tf.standardizeColOrder(dfDiff)
dfDiff
```

## Data Export

```python
tf.exportTemaData('goodExport.csv',df,includeNaN=True)
```

## Sample Data Visualization

```python
import altair as alt
chart3 = alt.Chart(tf.stripColUnit(dfDiff)).mark_line().encode(
    x = 'Time:Q',
    y = 'yVelocity1:Q',
    color=alt.value('green')
)
chart4 = alt.chart3 = alt.Chart(tf.stripColUnit(dfDiff)).mark_line().transform_window(
    rolling_mean='mean(yVelocity82)',
    frame=[-5, 5]
).encode(
    x = 'Time:Q',
    y = 'rolling_mean:Q',
    color=alt.value('purple')
)
chart4+chart3
```
